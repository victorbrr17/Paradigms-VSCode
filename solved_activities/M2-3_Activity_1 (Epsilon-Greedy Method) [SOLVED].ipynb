Plot the values of $\epsilon$ in the range 0 - 120000 using the following parameters:
- EPS_START = 1.0

```python
EPS_START = 1.0
EPS_DECAY = 0.999985
EPS_MIN = 0.2
```

<div class="alert alert-block alert-danger">

```python
def epsilon_decay(epsilon, decay, minimum):
    return max(epsilon * decay, minimum)
```

```python
epsilon = EPS_START
values = []
for i in range(120000):
    epsilon = epsilon_decay(epsilon, EPS_DECAY, EPS_MIN)
    values.append(epsilon)
```

```python
from matplotlib import pyplot as plt

plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(10,6))
plt.plot(values)
plt.axhline(y=EPS_MIN, color='r', linestyle='--')
plt.xlabel("Steps")
plt.ylabel("Epsilon")
plt.show()
```
